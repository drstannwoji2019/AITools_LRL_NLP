```python
# import NLTK
import nltk
```

```python
#import UDHR
from nltk.corpus import udhr
```

```python
#include Nigerian languages
languages = ['Edo', 'Ibibio_Efik', 'Hausa_Haoussa', 'NigerianPidginEnglish', 'Tiv']
cfd = nltk.ConditionalFreqDist((lang, len(word))
for lang in languages
for word in udhr.words(lang + "-Latin1"))
cfd.plot(cumulative=True)
```

```python
# Include other Nigerian Languages (UTF-8)
languages = ['Igbo', 'Yoruba']
cfd = nltk.ConditionalFreqDist((lang, len(word))
for lang in languages
for word in udhr.words(lang + "-UTF8"))
cfd.plot(cumulative=True)
```

```python
# Compare with high resource languages
languages = ['English', 'German_Deutsch', 'Spanish']
cfd = nltk.ConditionalFreqDist((lang, len(word))
for lang in languages
for word in udhr.words(lang + "-Latin1"))
cfd.plot(cumulative=True)
```
