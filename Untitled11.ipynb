```python
from transformers import pipeline

classifier = pipeline("sentiment-analysis")
classifier("For God so loved the world, He gave His only Begotten Son that whosoever believes in Him should not perish but have eternal life.")
```

```python
classifier(
    ["The wages of sin is death but the free gift of God is eternal life in Christ Jesus our Lord"]
)
```

```python
from transformers import pipeline

classifier = pipeline("zero-shot-classification")
classifier(
    "The proper interpretation of the Bible must be Christological and missiological. A holistic theological education is beneficial for people in business and ministry",
    candidate_labels=["education", "ministry", "theology"]
)
```
