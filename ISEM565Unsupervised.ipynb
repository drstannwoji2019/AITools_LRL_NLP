```python
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.mixture import GaussianMixture

# Load the Iris dataset
iris = load_iris()
data = pd.DataFrame(iris.data, columns=iris.feature_names)

# Step 2: K-means Clustering
scaler = StandardScaler()
scaled_data = scaler.fit_transform(data)
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans_labels = kmeans.fit_predict(scaled_data)

# Step 3: Expectation-Maximization (Gaussian Mixture Models)
gmm = GaussianMixture(n_components=3, random_state=42)
gmm_labels = gmm.fit_predict(scaled_data)

import matplotlib.pyplot as plt
import seaborn as sns

# Visualize K-means clustering
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
sns.scatterplot(x=data['sepal length (cm)'], y=data['sepal width (cm)'], hue=kmeans_labels, palette='viridis')
plt.title('K-means Clustering')

# Visualize GMM clustering
plt.subplot(1, 2, 2)
sns.scatterplot(x=data['sepal length (cm)'], y=data['sepal width (cm)'], hue=gmm_labels, palette='viridis')
plt.title('GMM Clustering')

plt.tight_layout()
plt.show()
```
